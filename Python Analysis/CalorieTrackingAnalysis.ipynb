```python
#Importing libraries

import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv('dailyActivity_mergedDataSet.csv')

# Display the DataFrame
df
```

# Cleaning Data :

```python
# Calculate the number of unique Id values in the DataFrame
unique_ids_count = df['Id'].nunique()

# Display the number of unique Id values
unique_ids_count
```

```python
# Selecting specific columns from the DataFrame
cols = ['Id', 'ActivityDate', 'TotalSteps', 'VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', 'SedentaryMinutes', 'Calories']

# Creating a new DataFrame with only the selected columns
df = df[cols]
```

```python
# Display the DataFrame
df
```

```python
# Rename the column 'ActivityDate' to 'Date' in the DataFrame df
df.rename(columns={'ActivityDate': 'Date'})
```

```python
# Calculate the total minutes by summing up VeryActiveMinutes, FairlyActiveMinutes, LightlyActiveMinutes, and SedentaryMinutes
df['TotalMinutes'] = df.VeryActiveMinutes + df.FairlyActiveMinutes + df.LightlyActiveMinutes + df.SedentaryMinutes
```

```python
# Calculate TotalHours by dividing TotalMinutes by 60 and rounding the result
df['TotalHours'] = round(df.TotalMinutes / 60)
```

```python
# Display the DataFrame
df
```

```python
# Display information about the DataFrame, including the data types and memory usage
df.info()
```

```python
# Convert the 'Date' column to datetime format using pandas
df.Date = pd.to_datetime(df.Date)
```

```python
# Display the DataFrame
df
```

```python
# Create a new column 'DayOfWeek' containing the day of the week corresponding to the dates in the 'Date' column
df['DayOfWeek'] = df.Date.dt.day_name()
```

```python
# Check for missing values in each column and sum them up
df.isnull().sum()
```

```python
# Count the number of duplicated rows in the DataFrame
df.duplicated().sum()
```

```python
# Generate descriptive statistics of the DataFrame, such as count, mean, standard deviation, minimum, maximum, etc.
df.describe()
```

# Visualization :

```python
# Set the size of the figure
plt.figure(figsize=(6, 4))

# Plot a histogram of the 'DayOfWeek' column with 7 bins representing each day of the week
# Set the color of the bars to orange and width to 0.6
plt.hist(df.DayOfWeek, bins=7, color='Orange', width=0.6)

# Add gridlines to the plot
plt.grid(True)

# Set the label for the x-axis
plt.xlabel('Day of Week')

# Set the label for the y-axis
plt.ylabel('Frequency')

# Set the title of the plot
plt.title('Frequency by Day of Week')

# Display the plot
plt.show()
```

```python
# Select only numeric columns from the DataFrame
numeric_df = df.select_dtypes(include=['number'])

# Compute the correlation matrix of the numeric columns
correlation_matrix = numeric_df.corr()

# Plot a heatmap of the correlation matrix using seaborn
# Set the color map to 'Oranges' for visualizing positive correlations
# Set annot=False to disable displaying correlation values on the heatmap
sns.heatmap(correlation_matrix, cmap='Oranges', annot=False)
```

```python
# Generate descriptive statistics of the DataFrame, such as count, mean, standard deviation, minimum, maximum, etc.
df.describe()
```

```python
# Set the size of the figure
plt.figure(figsize=(9, 6))

# Create a scatter plot of 'TotalSteps' versus 'Calories'
# Set the color of the points based on 'Calories'
plt.scatter(df.TotalSteps, df.Calories, c=df.Calories)

# Define median values for steps and calories
median_steps = 5986
median_calories = 2062

# Add a horizontal line at the median calories level
plt.axhline(median_calories, c='blue', label='Median of Calories')

# Add a vertical line at the median steps level
plt.axvline(median_steps, c='red', label='Median of Steps')

# Set the label for the x-axis
plt.xlabel('Steps')

# Set the label for the y-axis
plt.ylabel('Calories')

# Set the title of the plot
plt.title('Calories By Total Steps')

# Display legend to identify the median lines
plt.legend()

# Display the plot
plt.show()
```

```python
# Set the size of the figure
plt.figure(figsize=(9, 6))

# Create a scatter plot of 'TotalHours' versus 'Calories'
# Set the color of the points based on 'Calories'
plt.scatter(df.TotalHours, df.Calories, c=df.Calories)

# Define median values for hours and calories
median_hours = 24
median_calories = 2062

# Add a horizontal line at the median calories level
plt.axhline(median_calories, c='blue', label='Median of Calories')

# Add a vertical line at the median hours level
plt.axvline(median_hours, c='red', label='Median of Hours')

# Set the label for the x-axis
plt.xlabel('Hours')

# Set the label for the y-axis
plt.ylabel('Calories')

# Set the title of the plot
plt.title('Calories By Total Hours')

# Display legend to identify the median lines
plt.legend()

# Display the plot
plt.show()
```

```python
# Aggregate sums for each activity type
very_active_minutes_sum = df['VeryActiveMinutes'].sum()
fairly_active_minutes_sum = df['FairlyActiveMinutes'].sum()
lightly_active_minutes_sum = df['LightlyActiveMinutes'].sum()
sedentary_minutes_sum = df['SedentaryMinutes'].sum()

# Create a list of aggregated sums and corresponding labels
minutes = [very_active_minutes_sum, fairly_active_minutes_sum, lightly_active_minutes_sum, sedentary_minutes_sum]
labels = ['Very Active Minutes', 'Fairly Active Minutes', 'Lightly Active Minutes', 'Sedentary Minutes']

# Define a gradient of colors from orange to brown
colors = plt.cm.Oranges(np.linspace(0.2, 0.8, len(labels)))  # Adjust the range for desired shades

# Create the pie chart with custom colors
plt.pie(minutes, labels=labels, autopct='%1.1f%%', colors=colors)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

### ---------------------------------------------------------------------

### Summary of Data Analysis:

### Customer Activity by Day:

The bar chart highlights that customers are most active on Friday, Saturday, Sunday, and Tuesday compared to other days of the week. These days exhibit a higher frequency of customer engagement or usage.

### Correlation Analysis:

The heatmap chart reveals a strong positive correlation between total minutes spent in activities and sedentary minutes. Additionally, there is a notable correlation between total hours and sedentary minutes, indicating a relationship between the duration of activity and sedentary behavior.

### Scatter Plot Analysis:

The scatter plot chart demonstrates a positive correlation between calories burned and steps taken, suggesting that customers who take more steps tend to burn more calories.
Furthermore, there is a positive correlation between calories burned and hours spent, albeit weaker compared to the correlation between calories and steps.

### Pie Chart Analysis:

The pie chart illustrates the distribution of activity types, with sedentary minutes comprising the largest portion at 83.3%. In contrast, fairly active minutes represent the smallest portion at 1.1%.
These insights provide valuable information about customer behavior patterns and preferences, which can inform marketing strategies, product offerings, and initiatives aimed at promoting healthier lifestyles and increasing customer engagement.
